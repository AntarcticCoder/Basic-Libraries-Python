```python
#ส่งออกเอกสารหรือไฟล์นี้เเบบตารางออกไปเป็น csv
import pandas as pd
name = ["tuktuk", "dessert"]
price = [10,20]

nameS = pd.Series(name)
priceS = pd.Series(price)

frame = {
    "Name":nameS,
    "Price":priceS
}
df = pd.DataFrame(frame)
df.to_csv("product.csv") #ต้องระบุpath ที่จะใช้เก็บก่อน เเต่ถ้้าไม่ ก็เขียนชื่อไฟล์ไปเลย มันจะsave ไว้ที่ๆเดียวกับไฟลฺjupyter ที่ใช้เขียน เเต่ต้องsave ไฟล์jupyter  ก่อนด้วยนะ

#อันนี้คือุถ้าจะไม่เอา column เเละ index (header คือ column)
df.to_csv("productWithoutIndexandCols.csv", header=False, index=False)

```

```python
#การค้นหาข้อมูล
import pandas as pd
df = pd.read_excel("dataupdate.xlsx", "weather", index_col="Day")
df.Event.str.match("ฝน") #ค้นหาข้อมูลโดยคำขึ้นต้น (return True,False)
df[df.Event.str.match("ฝน")] #อันนี้จะreturn วันที่มีฝนตก ไม่ได้ return T,F
df[df.Event.str.contains("ตก")] #อันนี้หาค่าโดยคิดจากทุกคำในประโยค ไม่ได้ดูตัวขึ้นต้น
df[df.Event == "ฝนตก"] #ถ้าเเบบนี้คือต้องการจะหาข้อมูลที่เป็นคำนั้นๆเป้ะๆ เอาจริงๆก็เหมือน match เเหละ ไร้สาระ
```

```python
#การสอบถามข้อมูลเเบบมีเงื่อนไข
import pandas as pd
df = pd.read_excel("dataupdate.xlsx", "weather", index_col="Day")
df.Temperature<=20 #ข้อมูลจะออกมาเป็น T,F ว่า วันๆนั้น <=20 จิงป่าว
df[df.Temperature<=20] #ข้อมูลจะออกมาเป็นวันๆนั้นที่ <=20 เเบบเป็นข้อมูลออกมาเลย
df.loc[(df.Event == "ฝนตก") & (df.Temperature>=30)] #มากกว่า1 เงื่อนไข
#(ใน pandas ใช้ & เเทน and, ใช้ | เเทน or)
```

```python
#การใช้ | (or) เเบบลดรูป
import pandas as pd
df = pd.read_excel("dataupdate.xlsx", "weather", index_col="Day")
df.loc[(df.Temperature.isin([18,20,35]))] #เอาเเค่ 18 or 20 or 35 (ใช้เเทน or เขียนง่ายกว่า)
df.loc[(df.Temperature.isin([25,30]))] [df.Event=="แดดร้อน"] #isin and เงื่อนไชเพิ่มเติม
#(ส่งข้อมูลเข้าไปใน isin เป็น list นะเเจ้ะ)
#(ถ้าสมมติอยากได้เเค่ 35 ก็ใส่เเค่ 35 ใน isin ไม่ต้องใส่ == ไร้สาระ)
```

```python
#check data ว่ามีค่า null ไหม (ข้อมูลสูญหาย อาจพลาดอะไรสักอย่าง)
 #False = ข้อมูลครบ, True = มีข้อมูลหาย
import pandas as pd
df = pd.read_excel("Employee.xlsx", index_col="Name")
df.isnull() #หาว่ามีข้อมูลไหนเป็น nullไหม
df.isnull().any() #หาว่า columns ใดมีค่า null (ถ้ามี row หนึ่งใน column นั้นเป็น null มันจะบอกว่า True)
df.isnull().sum() #จะบอกว่าใน column นั้น มีช่องว่างกี่ตัว
```

```python
#เเทนที่ช่องว่าง(null) ด้วยข้อมูลใหม่
import pandas as pd
df = pd.read_excel("Employee.xlsx", index_col="Name")

#replace null by ค่าเฉลี่ย
df["Salary"].fillna(df["Salary"].mean(), inplace=True)
df #สรุปง่ายๆคือเข้าถึง column Salary เเล้วเข้าถึงช่องว่างด้วย fillna เเล้วก็ใส่ข้อมูลเข้าไป ซึ่งใส่ไปเป็นค่าเฉลี่ยของ Salary (49,666.666667)

#replace null by ตามที่คุณต้องการกำหนด
df = pd.read_excel("Employee.xlsx", index_col="Name")
df["Salary"].fillna(20000, inplace=True)
df

#replace null ทั้งหมดใน dataframe ยกเว้น index เป็นค่าก่อนหน้า (ควรตั้ง name เป็น index มันจะได้ไม่เอา name มาซ้ำ)
df = pd.read_excel("Employee.xlsx", index_col="Name")
df.fillna(method="pad", inplace=True)
df #จริงๆกำหนด column ก็ได้นะ เเบบอันก่อนๆ ["Salary"]

#replace null ทั้งหมดใน dataframe ยกเว้น index เป็นค่าถัดไป 
df = pd.read_excel("Employee.xlsx", index_col="Name")
df.fillna(method="bfill", inplace=True)
df
```

```python
#ลบเเถว ที่มี null
import pandas as pd
df = pd.read_excel("Employee.xlsx", index_col="Name")
df.dropna(inplace=True) #ลบเเถวทั้งหมดที่มีค่าว่าง
df

#ลบเเถวที่มีค่าว่าง ตาม column ที่ you กำหนด
df = pd.read_excel("Employee.xlsx", index_col="Name")
df.dropna(subset=["Age", "Job"]) #เช่น ถ้าcolumn Age, Job มีค่าว่าง ให้ลบ row นั้นทิ้ง
```
